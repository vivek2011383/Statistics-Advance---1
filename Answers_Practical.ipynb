```python
"""
2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
"""

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.arange(1, 7)  # Example: Rolling a fair die (values 1 to 6)
pmf_values = stats.randint.pmf(x, 1, 7)  # PMF of discrete uniform distribution

plt.bar(x, pmf_values, color='blue', alpha=0.7)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution')
plt.show()
```

```python
"""
3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
"""

from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

p = 0.5  # Probability of success
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]

for x, pdf in zip(x_values, pdf_values):
    print(f"P(X={x}) = {pdf}")
```

```python
"""
4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n, p = 10, 0.5  # Number of trials, Probability of success
data = binom.rvs(n, p, size=1000)

plt.hist(data, bins=np.arange(12)-0.5, density=True, alpha=0.7, color='blue')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution Histogram')
plt.show()
```

```python
"""
5. Create a Poisson distribution and visualize it using Python.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lambda_ = 4  # Mean number of events
x = np.arange(0, 15)
pmf_values = poisson.pmf(x, lambda_)

plt.bar(x, pmf_values, color='green', alpha=0.7)
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.show()
```

```python
"""
6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

x = np.arange(1, 7)  # Values of a fair die
cdf_values = randint.cdf(x, 1, 7)

plt.step(x, cdf_values, where='mid', color='red')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution')
plt.show()
```

```python
"""
19. Implement hypothesis testing using Z-statistics for a sample dataset.
"""

import numpy as np
from scipy import stats

sample_data = np.random.normal(50, 10, 30)
population_mean = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Z-Statistic: {z_statistic}, P-Value: {p_value}")
```

```python
"""
20. Create a confidence interval for a dataset using Python and interpret the result.
"""

import numpy as np
from scipy import stats

data = np.random.normal(50, 10, 30)
mean = np.mean(data)
std_err = stats.sem(data)
confidence = 0.95

interval = stats.t.interval(confidence, df=len(data)-1, loc=mean, scale=std_err)
print("Confidence Interval:", interval)
```

```python
"""
21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
"""

import numpy as np
from scipy import stats

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
std_err = stats.sem(data)
confidence_interval = stats.t.interval(0.95, len(data)-1, mean, std_err)

print("Mean:", mean)
print("95% Confidence Interval:", confidence_interval)
```

```python
"""
25. Write a Python program to compare the standard deviations of two datasets and visualize the difference.
"""

import numpy as np
import matplotlib.pyplot as plt

data1 = np.random.normal(50, 10, 1000)
data2 = np.random.normal(50, 20, 1000)

std1, std2 = np.std(data1), np.std(data2)
print("Standard Deviation of Dataset 1:", std1)
print("Standard Deviation of Dataset 2:", std2)

plt.hist(data1, bins=30, alpha=0.6, label="Dataset 1 (=10)")
plt.hist(data2, bins=30, alpha=0.6, label="Dataset 2 (=20)")
plt.legend()
plt.show()
```

```python
"""
26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
"""

import numpy as np
from scipy.stats import iqr

data = np.random.normal(50, 15, 1000)
range_value = np.ptp(data)  # Peak-to-Peak range
iqr_value = iqr(data)

print("Range:", range_value)
print("Interquartile Range (IQR):", iqr_value)
```
